```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
dfs = []
for f in Path("result_pickles").iterdir():
    fn = str(f)
    if "_r10" in fn or "_r1." in fn:
        print(fn)
        df = pd.read_pickle(fn)
        exp_name = fn.split("/")[-1].split(".")[0]
        df["dataset_rem_name"] = exp_name
        dfs.append(df)
        
all_ = pd.concat(dfs)
```

```python
model_rename = dict(
    zip(
        all_["model"].unique(),
        ["00 DCT", "01 DFT", "04 CP", "05 CP+S", "06 Tucker", "07 Tucker+S", "02 FB", "03 HW"]
    )
)
```

```python
all_['model'] = all_['model'].replace(model_rename)
```

```python
stat_cols = ["in_rmse", "in_mad", "out_rmse", "out_mad"]
all_by_pars = all_.groupby(
    ["dataset_rem_name", "model", "experiment_id", "parameters"]
)[stat_cols].mean()
```

```python
res0 = all_by_pars.groupby(["dataset_rem_name", "model"])[stat_cols].min()
```

```python
# res1 = res0[["out_rmse", "out_mad"]]
res1 = res0[["in_rmse", "in_mad"]]
```

```python
for stat in res1.columns:
    piv = res1[[stat]].pivot_table(
        index="dataset_rem_name",
        columns="model"
    )
    display(piv)
```
